```python
#한기대 사이트에서 특정 키워드로 자동 검색한 후 정보 추출하여 csv , xls 저장하기

#Step 1. 필요한 모듈을 로딩합니다
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
import time          

#Step 2. 사용자에게 검색 관련 정보들을 입력 받습니다.
print("=" *100)
print(" 이 크롤러는 한국기술교육대학교 정보 수집용 웹크롤러입니다.")
print("=" *100)
query_txt = input('1.수집할 자료의 키워드는 무엇입니까?(예: 장학금): ')
print("\n")

#Step 3. 크롬 드라이버 설정 및 웹 페이지 열기
s = Service("c:/py_temp/chromedriver.exe")
driver = webdriver.Chrome(service=s)

url = 'https://www.koreatech.ac.kr/'
driver.get(url)
time.sleep(5)
driver.maximize_window()

#Step 4. 자동으로 검색어 입력 후 조회하기
#돋보기 클릭하기
driver.find_element(By.XPATH,'//*[@id="header-wrap"]/a/span').click()
time.sleep(0.5)

element = driver.find_element(By.ID,'searchKeyword')
driver.find_element(By.ID,'searchKeyword').click( )
element.send_keys(query_txt)
element.send_keys("\n")

#Step 5.학위 논문 선택하기
driver.find_element(By.LINK_TEXT,'대학').click()
time.sleep(2)

#Step 6.Beautiful Soup 로 본문 내용만 추출하기
from bs4 import BeautifulSoup
html_1 = driver.page_source #현재 페이지의 전체 소스코드를 다 가져오기
soup_1 = BeautifulSoup(html_1, 'html.parser')

content_1 = soup_1.find_all('ul','searchrst-list campus-org')
content_2 = content_1[1].find_all('li')

for i in content_2 :
    print(i.get_text().replace("\n"," ").strip())
    print("\n")

# 상세내역 추출하여 xls , csv 로 저장하기
no2 = [ ]       #번호 저장용 리스트
title2 = [ ]    #제목 저장용 리스트
date2 = [ ]     #날짜 저장용 리스트
content2 = [ ]  #내용 저장용 리스트 

no = 1

#상세 내역 추출하기
for b in content_2 :
    #번호
    no2.append(no)
    print('1.번호:' , no)
    
    #제목
    title = b.find('a','ptit').get_text().strip()
    title2.append(title)
    print('2.제목:',title)
    
    #날짜
    date = b.find('p','date').get_text()
    date2.append(date)
    print('3.날짜:', date)
    
    #내용
    content = b.find('p','pcont').get_text().strip()
    content2.append(content)
    print('4.내용:',content)
    
    print()
    
    if no == 30 :
        break
        
    no += 1

#Step 7. 표준 출력 방향을 바꾸어 txt 파일에 저장하기
import pandas as pd
han = pd.DataFrame()
han['번호'] = no2
han['제목'] = title2
han['날짜'] = date2
han['내용'] = content2

fc_name = 'c:\\py_temp\\han.csv'
fx_name = 'c:\\py_temp\\han.xlsx'

han.to_csv(fc_name , encoding='utf-8-sig',index=False)
han.to_excel(fx_name , encoding='utf-8' , index=False)
print('요청하신 데이터 수집 작업이 정상적으로 완료되었습니다')
```
