```python
#연습문제 답지

#Line chart 연습문제

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.font_manager as fm
import matplotlib

# 한글 폰트 설정합니다
font_location = "c:\windows\Fonts\HYCYSM.TTF"
font_name = fm.FontProperties(fname = font_location).get_name()
matplotlib.rc('font' , family=font_name)

# 그래프 위치와 스타일을 설정합니다
fig = plt.figure(figsize=(10,5))
plt.style.use('ggplot')
ax = fig.add_subplot(1, 1, 1)

df = pd.read_csv("c:\\temp\\학생별과목별성적_3명.csv" , encoding="euc-kr")
print(df.head(6))
date = np.arange(1,7)
sub = ['국어','영어','수학','국사','화학','물리']

df_park = df[df['이름']=='박지영']
df_park_score = df_park['점수']
df_kim = df[df['이름']=='김태함']
df_kim_score = df_kim['점수']
df_kwon = df[df['이름']=='권혁진']
df_kwon_score = df_kwon['점수']

ax.plot(sub , df_park_score , label='박지영')
ax.plot(sub , df_kim_score , label='김태함')
ax.plot(sub , df_kwon_score , label='권혁진')

# 축 제목과 범례그리기
ax.set_title('학생별 과목별 성적', fontproperties = font_name)
ax.set_ylabel('점수(단위:점)',fontproperties = font_name)
ax.set_xlabel('과목', fontproperties = font_name)
ax.legend()
plt.show()
```

```python
#Bar chart 연습문제

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.font_manager as fm
import matplotlib

# 한글 폰트 설정합니다
font_location = "c:\windows\Fonts\HYCYSM.TTF"
font_name = fm.FontProperties(fname = font_location).get_name()
matplotlib.rc('font' , family=font_name)

df = pd.read_csv("c:\\temp\\홍길동시험점수.csv" , encoding="euc-kr")
df.head()

plt.style.use('ggplot')
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)

colors = ['red','blue','green']
labels = df['과목']
x = [1,2,3]
y = df['점수']

ax.set_title('홍길동 시험 점수', fontproperties = font_name)
ax.bar(x,y , tick_label = labels , color = colors)

#아래의 enumerate( ) 함수는 series 형태를 int 형태로 바꾸는함수임
for i , v in enumerate(y):
    ax.text(i+1, v+1.05, str(v)+'점', color='blue', fontweight='bold')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.font_manager as fm
import matplotlib

# 한글 폰트 설정합니다
font_location = "c:\windows\Fonts\HYCYSM.TTF"
font_name = fm.FontProperties(fname = font_location).get_name()
matplotlib.rc('font' , family=font_name)

# 그래프 위치와 스타일을 설정합니다
fig = plt.figure(figsize=(10,5))
plt.style.use('ggplot')
ax = fig.add_subplot(1, 1, 1)

df = pd.read_csv("c:\\temp\\1군전염병발병현황_년도별_2.csv",
                 encoding="euc-kr"  , index_col='년도')
df.head(10)

wt = np.array(range(len(df)))
w = 0.1
for i in df.columns :
    ax.bar(wt , df[i] , width=w , label=i)
    wt = wt + w
ax.set_xticks(np.array(range(len(df))))
ax.set_xticklabels(df.index,fontproperties=font_name)
ax.set_ylabel("발생건수",fontproperties=font_name)
ax.legend()
plt.show()
```

```python
#Folium 연습문제 1

import folium
import pandas as pd

df = pd.read_csv("c:\\temp\\경주주요관광지_LAT_LON.csv", encoding="euc-kr"  , index_col='no')
print(df)

dlon = df['lon']
dlat = df['lat']
dname=df['name']
print(dname[1])
```

```python
m4 = folium.Map(location=[dlat[1], dlon[1]], zoom_start=11)

for i in range(1,len(df)+1) :

    folium.Marker(location=[dlat[i], dlon[i]], popup=dname[i],
             icon=folium.Icon(icon='cloud')).add_to(m4)
m4
```
