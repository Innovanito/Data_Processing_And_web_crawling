```python
# pandas 연습문제 2
import openpyxl
wb = openpyxl.load_workbook("c:\\temp\\부품구입대장.xlsx")
sheet = wb["Sheet1"]

날짜2 = []
품목2 = []
수량2 = []
금액2 = []

for i in range(5, sheet.max_row + 1) :
    날짜 = sheet.cell(row=i,column=1).value
    날짜2.append(날짜)
    
    품목 = sheet.cell(row=i,column=3).value
    품목2.append(품목)
    
    수량 = sheet.cell(row=i,column=4).value
    수량2.append(수량)
    
    금액 = sheet.cell(row=i,column=6).value
    금액2.append(금액)
    
part = pd.DataFrame()
part['날짜'] = 날짜2
part['품목'] = 품목2
part['수량'] = 수량2
part['금액'] = 금액2

part
```

```python
# pandas 연습문제 3
part2 = part.loc[ (part['금액'] > 1000000)]

part2.sort_values(["금액"], ascending=[False])
```

```python
# pandas 연습문제 4
part2 = part.loc[ (part['금액'] > 1000000,['품목','금액'])]

part2.sort_values(["금액"], ascending=[False])
```

```python
# pandas 연습문제 5
# 부품구입대장에서 품명이 '백관' 일 경우 코드 컬럼에 'A1' , 
# '절연판' 일경우 'B1' , 실리콘일 경우 'C1' 값을 입력하도록
# 코드를 작성하세요

import pandas as pd
import openpyxl
wb = openpyxl.load_workbook("c:\\temp\\부품구입대장.xlsx")
sheet = wb["Sheet1"]

날짜2 = []
품목2 = []
수량2 = []
금액2 = []
코드2 = []

for i in range(5, sheet.max_row + 1) :
    날짜 = sheet.cell(row=i,column=1).value
    날짜2.append(날짜)
    
    품목 = sheet.cell(row=i,column=3).value
    품목2.append(품목)
    if 품목=='백관' :
        코드2.append('A1')
    elif 품목=='절연판' :
        코드2.append('B1')
    elif 품목=='실리콘' :
        코드2.append('C1')
    else :
        코드2.append('')
    
    수량 = sheet.cell(row=i,column=4).value
    수량2.append(수량)
    
    금액 = sheet.cell(row=i,column=6).value
    금액2.append(금액)
    
part = pd.DataFrame()
part['날짜'] = 날짜2
part['품목'] = 품목2
part['수량'] = 수량2
part['금액'] = 금액2
part['코드'] = 코드2

part
```
